```python
# Import modules
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Define data filepath
pathname = '/Users/jryan4/Dropbox (University of Oregon)/Teaching/geospatial-data-science/data/lab2/'

# Read data
df = gpd.read_file(pathname + 'or_1992-2018.shp')
```

* a) Which **county** had the most **human** caused wildfires **>50 acres**? (HINT: use the `FIPS_NAME` column)

```python
df_human_50 = df[(df['FIRE_SIZE'] > 50) & (df['NWCG_CAUSE'] == 'Human')].copy()
```

```python
df_human_50['FIPS_NAME'].describe()
```

```python
df_human_50.groupby(by='FIPS_NAME').count().sort_values(by='OBJECTID', ascending=False)
```

* b) Which **month** had the most **natural** caused wildfires **>100 acres**?

```python
df_natural_100 = df[(df['FIRE_SIZE'] > 100) & (df['NWCG_CAUSE'] == 'Natural')].copy()
```

```python
datetime = pd.to_datetime(df_natural_100['DISCOVERY_'], format='%Y/%m/%d %H:%M:%S.%f')
df_natural_100['datetime'] = datetime
df_natural_100_count = df_natural_100.groupby(df_natural_100['datetime'].dt.month).count()
```

```python
df_natural_100_count.sort_values(by='OBJECTID', ascending=False)
```

* c) How many fires **>200 acres** have an **undetermined** cause (e.g. `Missing data/not specified/undetermined`?
